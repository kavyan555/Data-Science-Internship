task 1:

analyse the sales data:

total sales per region

average sales per month

best performing region

```python
import numpy as np

# Rows = Months, Columns = Regions
# Regions: North, South, East

sales = np.array([
    [200, 150, 300],   # Month 1
    [250, 180, 320],   # Month 2
    [300, 200, 350],   # Month 3
    [280, 210, 400]    # Month 4
])

print("Sales Data:\n", sales)
```

```python
total_region_sales = np.sum(sales, axis=0)
print("Total Sales Per Region:", total_region_sales)
```

```python
avg_month_sales = np.mean(sales, axis=1)
print("Average Sales Per Month:", avg_month_sales)
```

```python
best_region_index = np.argmax(total_region_sales)

regions = ["North", "South", "East"]

print("Best Performing Region:", regions[best_region_index])
```

task 2:

student ranking system(sorting)

```python
import numpy as np

students = np.array([
    (101, "Ravi", 85),
    (102, "Anita", 92),
    (103, "Kiran", 78),
    (104, "Meena", 95),
    (105, "Arjun", 88)
], dtype=[('id', 'i4'), ('name', 'U10'), ('marks', 'i4')])

print("Student Data:\n", students)
```

```python
sorted_students = np.sort(students, order='marks')[::-1]

print("Students Ranked by Marks:\n", sorted_students)
```

```python
print("\nStudent Ranking List:")
rank = 1
for student in sorted_students:
    print("Rank", rank, ":", student['name'], "- Marks:", student['marks'])
    rank += 1
```

task3:

production quality check

```python
import numpy as np

# Columns: ProductID, Produced Units, Defective Units
production = np.array([
    [101, 500, 20],
    [102, 600, 30],
    [103, 550, 25],
    [104, 700, 50],
    [105, 650, 15]
])

print("Production Data:\n", production)
```

```python
produced = production[:, 1]
defective = production[:, 2]
```

```python
defect_rate = (defective / produced) * 100
print("Defect Percentage:\n", defect_rate)
```

```python
avg_defect = np.mean(defect_rate)
print("Average Defect Rate:", avg_defect)
```

```python
poor_quality = production[defect_rate > avg_defect]

print("Poor Quality Products:\n", poor_quality)
```

```python
best_index = np.argmin(defect_rate)

print("Best Quality Product:", production[best_index])
```

task 4:

broadcasting in pricing system

```python
import numpy as np

# Product prices in different regions
#columns=regions
prices = np.array([
    [100, 110, 120],   # Product 1
    [200, 210, 220],   # Product 2
    [300, 310, 320]    # Product 3
])

print("Original Prices:\n", prices)
```

Apply Discount Using Broadcasting

Suppose each region has a different discount

Region discounts:

North → 10
South → 15
East → 20

```python
discount = np.array([10, 15, 20])

new_prices = prices - discount

print("Prices After Discount:\n", new_prices)
```

```python
#Add Tax Using Broadcasting

taxed_prices = new_prices * 1.05

print("Final Prices After Tax:\n", taxed_prices)
```

task 5:
 create one data with numpy:

 customer id, age, purchase_amount region

 1.find the total customers

 2.extract only purchase column with slicing technique

 3.find the average purchase amount

 4.find the maximum and minimum purchase

 5.find the customers above average purchase

```python
import numpy as np

# Creating dataset
# Columns: ID, Age, Purchase, Region
data = np.array([
    [101, 25, 500, 1],
    [102, 30, 700, 2],
    [103, 22, 300, 1],
    [104, 35, 900, 3],
    [105, 28, 650, 2],
    [106, 40, 1200, 3]
])

print("Customer Data:\n", data)
```

```python
total_customers = data.shape[0]
print("Total Customers:", total_customers)
```

```python
purchase = data[:, 2]
print("Purchase Amount Column:", purchase)
```

```python
avg_purchase = np.mean(purchase)
print("Average Purchase Amount:", avg_purchase)
```

```python
max_purchase = np.max(purchase)
min_purchase = np.min(purchase)

print("Maximum Purchase:", max_purchase)
print("Minimum Purchase:", min_purchase)
```

```python
above_avg_customers = data[purchase > avg_purchase]

print("Customers Above Average Purchase:\n", above_avg_customers)
```

task 6:

create one dataset:
employee id,
employee name,
age,
salary

find the total salary of all employees

find the minimum salary of the employees

find the maximum salary of the employees

find the average age of employees

find the total number of employees

find the correlation between age and salary


```python
import numpy as np

# Creating structured dataset
employees = np.array([
    (101, "Ravi", 25, 30000),
    (102, "Anita", 30, 45000),
    (103, "Kiran", 28, 40000),
    (104, "Meena", 35, 55000),
    (105, "Arjun", 32, 50000)
], dtype=[('id', 'i4'), ('name', 'U10'), ('age', 'i4'), ('salary', 'i4')])

print(employees)
```

```python
total_salary = np.sum(employees['salary'])
print("Total Salary:", total_salary)
```

```python
min_salary = np.min(employees['salary'])
print("Minimum Salary:", min_salary)
```

```python
max_salary = np.max(employees['salary'])
print("Maximum Salary:", max_salary)
```

```python
avg_age = np.mean(employees['age'])
print("Average Age:", avg_age)
```

```python
total_employees = len(employees)
print("Total Employees:", total_employees)
```

```python
correlation = np.corrcoef(employees['age'], employees['salary'])
print("Correlation Matrix:\n", correlation)

print("Correlation between Age & Salary:", correlation[0,1])
```

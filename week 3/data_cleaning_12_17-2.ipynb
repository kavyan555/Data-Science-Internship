data cleaning:

it is the process of identifying and removing any missing,duplicate or irrelevant data

missing value/null values

duplicates

wrong data

wrong format

data cleaning is very important in exploratory data analysiss because to take right actions which we get from
the insights of the data

benefits of data cleaning:
- error free
- data quality
- accuracy & efficient
- maintain data consistency


1.Handling missing values:

we needd to remove missing values to avoid the erros

#check missing values
df.isnull().sum()

#drop missing values
df.dropna()

#drop specific column
df.dropna(subset=["Age"])

#fill missing values
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["city"].fillna(df["city"].mode(),inplace=True)
df["salary"].fillna(df["salary"].median(),inplace=True)

2.Remove duplicates
remove the repeated rows

#to remove duplicates
df.drop_duplicates(inplace=True)

#to check duplicates
df.duplicated().sum()


3.wrong data
irrelevant information
example age=250(wrong data)

df[df["Age"]>=100

4.data type correction
df.dtypes

df["Age"]=df["Age"].astype(int)


5.wrong format
ex:joining data

df["joining_date"]=pd.to_datetime(df["joining_date"])

df["gender"]=df["gender"].str.lower()

```python
df=pd.DataFrame(data)
```

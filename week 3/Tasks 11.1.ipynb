task 1:

create a dataframe where it consists of columns like name,dept,salary


filter by department

group by department

```python
import pandas as pd

data = {
    "Name": ["Asha", "Ravi", "Kiran", "Meena", "Arjun", "Priya"],
    "Dept": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Salary": [35000, 60000, 55000, 50000, 40000, 52000]
}

df = pd.DataFrame(data)
print(df)
```

```python
it_employees = df[df["Dept"] == "IT"]
print(it_employees)
```

```python
grouped = df.groupby("Dept")["Salary"].mean()
print(grouped)
```

task 2:

create a dataframe where the columns like product id,product name,quantity,and price

find the top 5 products based on the total price

```python
products = {
    "Product_ID": [101, 102, 103, 104, 105, 106, 107],
    "Product_Name": ["Laptop", "Mouse", "Keyboard", "Monitor", "Printer", "Tablet", "Headphones"],
    "Quantity": [5, 20, 15, 7, 4, 6, 10],
    "Price": [50000, 500, 1500, 12000, 8000, 20000, 3000]
}

prod_df = pd.DataFrame(products)
print(prod_df)
```

```python
prod_df["Total_Price"] = prod_df["Quantity"] * prod_df["Price"]

top5 = prod_df.sort_values(by="Total_Price", ascending=False).head(5)
print(top5)
```

task 3:

create a two datasets customer details and order details

merge the two datsets on the basis of customer_id

```python
customers = {
    "Customer_ID": [1, 2, 3, 4],
    "Customer_Name": ["Asha", "Ravi", "Meena", "Kiran"],
    "City": ["Bangalore", "Chennai", "Hyderabad", "Delhi"]
}

cust_df = pd.DataFrame(customers)
```

```python
cust_df
```

```python
orders = {
    "Order_ID": [1001, 1002, 1003, 1004],
    "Customer_ID": [1, 2, 1, 3],
    "Amount": [5000, 7000, 2000, 6500]
}

order_df = pd.DataFrame(orders)
```

```python
order_df
```

```python
merged_df = pd.merge(cust_df, order_df, on="Customer_ID", how="inner")
print(merged_df)
```

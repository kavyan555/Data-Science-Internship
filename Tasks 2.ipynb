```python
# Employee Salary Calculator

basic_salary = 30000
bonus_percent = 10
tax_percent = 5

# calculate bonus and tax
bonus_amount = (basic_salary * bonus_percent) / 100
tax_amount = (basic_salary * tax_percent) / 100

# final salary
final_salary = basic_salary + bonus_amount - tax_amount

print("Basic Salary:", basic_salary)
print("Bonus Amount:", bonus_amount)
print("Tax Amount:", tax_amount)
print("Final Monthly Salary:", final_salary)
```

```python
# E-Commerce Discount Logic

cart_value = float(input("Enter total cart value: "))

if cart_value >= 10000:
    discount = cart_value * 0.20
elif cart_value >= 5000:
    discount = cart_value * 0.10
else:
    discount = 0

final_amount = cart_value - discount

print("Cart Value:", cart_value)
print("Discount Applied:", discount)
print("Final Payable Amount:", final_amount)
```

```python
# API Response Status Checker

status_code = int(input("Enter API status code: "))

if status_code == 200:
    message = "Success: Request completed successfully"
elif status_code == 201:
    message = "Success: Resource created successfully"
elif status_code == 400:
    message = "Client Error: Bad request"
elif status_code == 401:
    message = "Client Error: Unauthorized access"
elif status_code == 404:
    message = "Client Error: Resource not found"
elif status_code == 500:
    message = "Server Error: Internal server error"
else:
    message = "Unknown Status Code"

print(message)
```

```python
# Internship Eligibility Checker using Dictionary

student = {
    "name": "Kavya",
    "cgpa": 9.6,
    "year": 4,
    "skill_score": 72
}

if student["cgpa"] >= 7.0 and student["year"] >= 3 and student["skill_score"] >= 60:
    print("Student Name:", student["name"])
    print("Status: Eligible for Internship")
else:
    print("Student Name:", student["name"])
    print("Status: Not Eligible for Internship")
```

```python
# E-Commerce Discount Engine

order = {
    "customer_name": "Kavya",
    "membership": "premium",   # premium / regular
    "cart_value": 8500
}

if order["membership"] == "premium":
    if order["cart_value"] >= 10000:
        discount = order["cart_value"] * 0.25
    elif order["cart_value"] >= 5000:
        discount = order["cart_value"] * 0.15
    else:
        discount = 0
else:  # regular member
    if order["cart_value"] >= 10000:
        discount = order["cart_value"] * 0.15
    elif order["cart_value"] >= 5000:
        discount = order["cart_value"] * 0.10
    else:
        discount = 0

final_amount = order["cart_value"] - discount

print("Customer Name:", order["customer_name"])
print("Membership:", order["membership"])
print("Cart Value:", order["cart_value"])
print("Discount Applied:", discount)
print("Final Payable Amount:", final_amount)
```

```python
# Bank Account Balance Validation

account = {
    "account_holder": "Kavya",
    "account_number": "1234567890",
    "balance": 12000
}

withdraw_amount = 4500

if withdraw_amount <= account["balance"]:
    account["balance"] -= withdraw_amount
    print("Withdrawal Successful")
    print("Amount Withdrawn:", withdraw_amount)
    print("Remaining Balance:", account["balance"])
else:
    print("Insufficient Balance")
    print("Available Balance:", account["balance"])
```

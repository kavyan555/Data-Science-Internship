task 1:

analyse the sales data:

total sales per region

average sales per month

best performing region

```python
import numpy as np

# Rows = Months, Columns = Regions
# Regions: North, South, East

sales = np.array([
    [200, 150, 300],   # Month 1
    [250, 180, 320],   # Month 2
    [300, 200, 350],   # Month 3
    [280, 210, 400]    # Month 4
])

print("Sales Data:\n", sales)
```

```python
total_region_sales = np.sum(sales, axis=0)
print("Total Sales Per Region:", total_region_sales)
```

```python
avg_month_sales = np.mean(sales, axis=1)
print("Average Sales Per Month:", avg_month_sales)
```

```python
best_region_index = np.argmax(total_region_sales)

regions = ["North", "South", "East"]

print("Best Performing Region:", regions[best_region_index])
```

task 2:

student ranking system(sorting)

```python
import numpy as np

students = np.array([
    (101, "Ravi", 85),
    (102, "Anita", 92),
    (103, "Kiran", 78),
    (104, "Meena", 95),
    (105, "Arjun", 88)
], dtype=[('id', 'i4'), ('name', 'U10'), ('marks', 'i4')])

print("Student Data:\n", students)
```

```python
sorted_students = np.sort(students, order='marks')[::-1]

print("Students Ranked by Marks:\n", sorted_students)
```

```python
print("\nStudent Ranking List:")
rank = 1
for student in sorted_students:
    print("Rank", rank, ":", student['name'], "- Marks:", student['marks'])
    rank += 1
```

task3:

production quality check

```python
import numpy as np

# Columns: ProductID, Produced Units, Defective Units
production = np.array([
    [101, 500, 20],
    [102, 600, 30],
    [103, 550, 25],
    [104, 700, 50],
    [105, 650, 15]
])

print("Production Data:\n", production)
```

```python
produced = production[:, 1]
defective = production[:, 2]
```

```python
defect_rate = (defective / produced) * 100
print("Defect Percentage:\n", defect_rate)
```

```python
avg_defect = np.mean(defect_rate)
print("Average Defect Rate:", avg_defect)
```

```python
poor_quality = production[defect_rate > avg_defect]

print("Poor Quality Products:\n", poor_quality)
```

```python
best_index = np.argmin(defect_rate)

print("Best Quality Product:", production[best_index])
```

task 4:

broadcasting in pricing system

```python
import numpy as np

# Product prices in different regions
#columns=regions
prices = np.array([
    [100, 110, 120],   # Product 1
    [200, 210, 220],   # Product 2
    [300, 310, 320]    # Product 3
])

print("Original Prices:\n", prices)
```

Apply Discount Using Broadcasting

Suppose each region has a different discount

Region discounts:

North → 10
South → 15
East → 20

```python
discount = np.array([10, 15, 20])

new_prices = prices - discount

print("Prices After Discount:\n", new_prices)
```

```python
#Add Tax Using Broadcasting

taxed_prices = new_prices * 1.05

print("Final Prices After Tax:\n", taxed_prices)
```

1. What are broadcasting rules in NumPy?

Broadcasting is a technique in NumPy that allows arithmetic operations between arrays of different shapes without manually reshaping them.

Broadcasting Rules:

- NumPy compares array shapes from right to left.

- Two dimensions are compatible when:

They are equal, OR

One of them is 1.

- If dimensions are not compatible, broadcasting fails.

Example:

A shape = (3, 4)

B shape = (1, 4)

Result: B is stretched to match A → operation possible.

2. Why is broadcasting important in Machine Learning?

Broadcasting is important because:

- It avoids writing loops, making computations faster.

- It reduces memory usage.

- It simplifies mathematical operations on datasets.

- It is heavily used in vectorized ML calculations like:

 Feature scaling

Matrix operations

Neural network computations

4. Difference between rand() and randint()

| Feature | rand()                           | randint()               |
| ------- | -------------------------------- | ----------------------- |
| Type    | Float values                     | Integer values          |
| Range   | 0 to 1                           | User-defined range      |
| Usage   | Generates decimal random numbers | Generates whole numbers |


Example:

```python
import numpy as np
np.random.rand(3)       # [0.34, 0.78, 0.12]
np.random.randint(1,10,3)  # [3, 7, 2]
```

9. Difference between copy and view

| Feature        | Copy               | View               |
| -------------- | ------------------ | ------------------ |
| Memory         | New memory created | Shares same memory |
| Changes affect | Only copied array  | Both arrays        |
| Method         | arr.copy()         | arr.view()         |


10. Difference between flatten() and ravel()

Both convert multi-dimensional arrays into 1D arrays.

| Feature      | flatten()                | ravel()                    |
| ------------ | ------------------------ | -------------------------- |
| Memory       | Creates new copy         | Returns view when possible |
| Speed        | Slower                   | Faster                     |
| Modification | Does NOT affect original | May affect original        |


11. Difference between copy and shallow copy

i. Copy (Deep Copy):

- Creates completely independent data.

- Changes do not affect original.

ii. Shallow Copy:

- Copies references only.

- Changes in nested objects affect original.

In NumPy:

- copy() → deep copy

- view() → shallow copy

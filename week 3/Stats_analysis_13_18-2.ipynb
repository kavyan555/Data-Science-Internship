data:information of facts(structure and unstructure data)

types of data:

1.qualitative:emotions,feelings---categorical data

2.quantitative:measurements,height,weight----continuous data

1.qualitative divides into two types:

- nominal data: categories without any order

ex:color,gender etc

- ordinal data: with some exact meaning

ex:hot,cold,sad,happy,low,high etc

2.quantitative data divides into two types

- interval: it shows the difference between the values

ex: 50 degrees

- ratio: which will be measured with some range

we can collect the data from different sources like:

1.primary source :collecting the data within ourselves

2.secondary source:kaggle,uci machine learning,government websites

correlation analysis:
it is used to show the relationship between independent and dependent variables

the values lies between 0 and 1,-1 to +1

if our value is close to 1 then it is a strong correlation

example:0.8,0.989,0.56789

if our value  is far to 1 then it is a weak correlation

example:0.0089,0.0076

strong positive correlation(0.98)

strong negative correlation(-0.76)

weak positive correlation(0.00007)

weak negative correlation(-0.008857)

regression analysis:
it shows the relation between input and output to predict the future values

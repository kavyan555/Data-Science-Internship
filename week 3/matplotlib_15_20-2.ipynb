A telecom company where the company wants some of the insights:

identifying churn trends

detect high risk customers

understand the revenue impact

support retention strategy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

n=500

data=pd.DataFrame({
    "Tenure":np.random.randint(1,72,n),
    "Monthly_charges":np.random.normal(600,150,n),
    "contract_type":np.random.choice(["month-to-month","oneyear","twoyear"],n),
    "churn":np.random.choice([0,1],n)
})
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

# INSIGHT 1  Identifying Churn Trends

```python
 #Churn Rate
print("Churn Rate:", data["churn"].mean()*100)

# Churn trends
print(pd.crosstab(data["contract_type"], data["churn"], normalize="index")*100)

# High Risk Customers
high_risk = data[(data["Tenure"] < 12) &
                 (data["Monthly_charges"] > data["Monthly_charges"].mean()) &
                 (data["contract_type"] == "month-to-month")]

print("High Risk Customers:", len(high_risk))
```

```python
plt.figure()
data["churn"].value_counts().plot(kind="bar")
plt.title("Overall Churn Distribution")
plt.xlabel("Churn (0 = Stayed, 1 = Left)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure()
pd.crosstab(data["contract_type"], data["churn"]).plot(kind="bar")
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Customer Count")
plt.show()
```

```python
# Create tenure groups
data["Tenure_Group"] = pd.cut(
    data["Tenure"],
    bins=[0,12,24,36,48,60,72],
    labels=["0-1yr","1-2yr","2-3yr","3-4yr","4-5yr","5-6yr"]
)

# Calculate churn rate
churn_trend = data.groupby("Tenure_Group")["churn"].mean()

# Line chart
plt.figure()
churn_trend.plot(kind="line", marker="o")
plt.title("Churn Trend by Customer Tenure")
plt.xlabel("Tenure Group")
plt.ylabel("Churn Rate")
plt.show()
```

# Detect High Risk Customers (Bar Chart)

```python
high_risk = data[(data["Tenure"] < 12)]

plt.figure()
high_risk.groupby("contract_type")["churn"].mean().plot(kind="bar")
plt.title("High Risk Customers by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

# Understand Revenue Impact (Bar Chart)

```python
plt.figure()
data.groupby("churn")["Monthly_charges"].sum().plot(kind="bar")
plt.title("Revenue Impact of Churn")
plt.xlabel("Churn")
plt.ylabel("Total Revenue")
plt.show()
```

# Support Retention Strategy (Line + Bar)

```python
retention = data.groupby("contract_type")["churn"].mean()

plt.figure()
retention.plot(kind="bar")
plt.title("Retention Strategy Insight")
plt.xlabel("Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

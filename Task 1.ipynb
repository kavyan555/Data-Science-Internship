Task 2: Clean csv raw log data


```python
import pandas as pd

raw_data = {
    'name': [' Shreya', 'Suhas', 'shreyas', 'Swetha', 'Suhas', None],
    'age': [25, 26, None, 24, 26, 23],
    'location': [' Hyderabad ', 'Noida', 'Gurugram', None, 'Noida', ' Mumbai ']
}

raw_df = pd.DataFrame(raw_data)

# save raw logs
raw_df.to_csv("raw_logs.csv", index=False)
print("Raw log file created")
```

```python
df = pd.read_csv("raw_logs.csv")
print(df)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
# remove duplicates
df = df.drop_duplicates()

# handle missing values
df['age'] = df['age'].fillna(df['age'].mean())
df['name'] = df['name'].fillna('unknown')
df['location'] = df['location'].fillna('unknown')

# clean text
df['name'] = df['name'].str.strip().str.lower()
df['location'] = df['location'].str.strip().str.lower()

# fix datatype
df['age'] = df['age'].astype(int)
```

```python
df.to_csv("cleaned_raw_logs.csv", index=False)
print("Cleaned raw logs saved successfully")
```

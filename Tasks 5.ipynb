4.create a rectangle class with length and width
write methods to calculate area and perimeter

```python
# Create Rectangle class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create rectangle object
rect = Rectangle(10, 5)

# Print results
print("Area of Rectangle:", rect.area())
print("Perimeter of Rectangle:", rect.perimeter())
```

6.create a bankaccount class with private balance
add method like
deposit
withdraw
get balance

```python
# Create BankAccount class
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

# Create object
account = BankAccount(5000)

# Perform operations
account.deposit(2000)
account.withdraw(3000)

# Print balance
print("Current Balance:", account.get_balance())
```

8.create a class atm that hides the pin and allow the transaction only after validation

```python
# Create ATM class
class ATM:
    def __init__(self, pin, balance):
        self.__pin = pin          # private PIN
        self.__balance = balance  # private balance

    def validate_pin(self, entered_pin):
        return entered_pin == self.__pin

    def withdraw(self, entered_pin, amount):
        if self.validate_pin(entered_pin):
            if amount <= self.__balance:
                self.__balance -= amount
                print("Withdrawal successful")
            else:
                print("Insufficient balance")
        else:
            print("Invalid PIN")

    def get_balance(self, entered_pin):
        if self.validate_pin(entered_pin):
            return self.__balance
        else:
            return "Invalid PIN"

# Create ATM object
atm = ATM(1234, 10000)

# Perform transactions
atm.withdraw(1234, 3000)
print("Balance:", atm.get_balance(1234))
```

1.build a library management system using
class
inheritance
encapsulation

```python
# Base class
class LibraryItem:
    def __init__(self, title):
        self.title = title

    def display_info(self):
        print("Title:", self.title)


# Derived class (Inheritance)
class Book(LibraryItem):
    def __init__(self, title, author, copies):
        super().__init__(title)
        self.author = author
        self.__copies = copies   # private variable (Encapsulation)

    # Method to issue a book
    def issue_book(self):
        if self.__copies > 0:
            self.__copies -= 1
            print("Book issued successfully")
        else:
            print("No copies available")

    # Method to return a book
    def return_book(self):
        self.__copies += 1
        print("Book returned successfully")

    # Getter method to access private data
    def get_copies(self):
        return self.__copies

    def display_info(self):
        print("Title  :", self.title)
        print("Author :", self.author)
        print("Copies :", self.__copies)
        print("------------------------")


# Create object
book1 = Book("Python Programming", "Guido van Rossum", 3)

# Operations
book1.display_info()
book1.issue_book()
book1.issue_book()
book1.return_book()

print("Available Copies:", book1.get_copies())
```

4.employee management system

tasks:
create employee base class
private salary variable
child classes:manager,intern
override calculate_salary
store multipple employees in a list
display the pay roll

```python
# Base class
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary   # private variable

    def calculate_salary(self):
        return self.__salary

    def get_salary(self):
        return self.__salary


# Child class: Manager
class Manager(Employee):
    def calculate_salary(self):
        # Manager gets bonus
        return self.get_salary() + 10000


# Child class: Intern
class Intern(Employee):
    def calculate_salary(self):
        # Intern gets stipend (fixed)
        return self.get_salary()


# Store multiple employees in a list
employees = [
    Manager("Amit", 50000),
    Intern("Neha", 15000),
    Manager("Ravi", 60000),
    Intern("Kiran", 12000)
]

# Display payroll
print("------ PAYROLL DETAILS ------")
for emp in employees:
    print("Name   :", emp.name)
    print("Salary :", emp.calculate_salary())
    print("-----------------------------")
```

```python
class product:
    def __init__(self,name,price):
        self.name=name
        self.price=price

    def get_price(self):
        return self.price

    def get_details(self):
        return f" {self,name}- RS{self.price}"
```

```python
class Electronics(product):
    def __init__(self,name,price,waranty_years):
        super().__init__(name,price)
        self.waranty_years=waranty_years

    def get_details(self):
        return f"Electronics:{self.name}, price: Rs {self.price}, waranty:{self.waranty_years} years"
```

```python
class clothing(product):
    def __init__(self,name,price,size):
        super().__init__(name,price)
        self.size=size

    def get_details(self):
        return f"clothing:{self.name}, price : Rs {self.price}, size:{self.size}"
```

```python
class cart:
    def __init__(self):
        self.items=[]

    def add_product(self,product):
        self.items.append(product)
        print(f"Added to cart:{product.name}")

    def calculate_total(self):
        return sum(item.get_price() for item in self.items)

    def show_cart(self):
       print("\n cart items:")
       for item in self.items:
           print(item.get_details())
```

```python
class payment:
    def pay(self,amount):
        raise NotImplementError("payment method not selected")
```

```python
class creditcardpayment(payment):
    def pay(self,amount):
        print(f"paid {amount} through credit card")
```

```python
class upipayment(payment):
    def pay(self,amount):
        print( f"paid {amount} using upi")
```

```python
class Bill:
    def __init__(self, cart, payment_method):
        self.cart = cart
        self.payment_method = payment_method

    def generate_bill(self):
        self.cart.show_cart()
        total = self.cart.calculate_total()

        print("\nBill Summary")
        print(f"Total Amount: Rs {total}")

        self.payment_method.pay(total)
        print("Order placed successfully")
```

```python
import numpy as np
import pandas as pd
np.random.seed(42)
n=1000

data=pd.DataFrame({
    "Customer_id":np.arange(1,n+1),
    "Age":np.random.normal(35,10,n).astype(int),
    "Gender":np.random.choice(["Male","Female"],n),
    "Income":np.random.normal(50000,15000,n),
    "Spending_Score":np.random.uniform(1,100,n),
    "Region":np.random.choice(["North","South","East","West"],n),
    "Ad_Spend":np.random.normal(20000,5000,n),
    "Website_Time":np.random.exponential(5,n),
    "Purchased":np.random.choice([0,1],n),
    "Satisfaction_Score":np.random.normal(7,1.5,n),
    "Delivery_Time":np.random.poisson(3,n),
    "Sales":np.random.normal(300,80,n)
})
```

8.how to detect the outliers?

```python
import seaborn as sns
```

```python
sns.boxplot(data['Sales'])
```

```python
old_shape=data.shape
```

```python
old_shape
```

```python
# to remove outliers using IQR
Q1=data['Sales'].quantile(0.25)
Q3=data['Sales'].quantile(0.75)

IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

data_cleaned=data[(data['Sales']>=lower_bound)&(data['Sales']<=upper_bound)]

new_shape=data_cleaned.shape
```

```python
new_shape
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.boxplot(data['Sales'])
plt.title("Boxplot of sales before removing outliers")
plt.ylabel("Sales")
plt.show()


plt.figure()
plt.boxplot(data_cleaned['Sales'])
plt.title("Boxplot of sales after removing outliers")
plt.ylabel("Sales")
plt.show()
```

9.standardize the sales?

(Z-Score)

Converts to mean=0, std=1

```python
data["Sales_Standardized"] = (data["Sales"] - data["Sales"].mean()) / data["Sales"].std()

data["Sales_Standardized"].head()
```

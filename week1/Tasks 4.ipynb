5. create a custom exception rules for the atm withdrawal

```python
class ATMError(Exception):
    pass
```

```python
def withdraw(balance, amount):
    if amount % 100 != 0:
        raise ATMError("Amount must be in multiples of 100")

    if amount > balance:
        raise ATMError("Insufficient balance")

    if balance - amount < 1000:
        raise ATMError("Minimum balance of 1000 must be maintained")

    return balance - amount
```

```python
try:
    balance = 5000
    amount = int(input("Enter withdrawal amount: "))
    balance = withdraw(balance, amount)
    print(" Withdrawal successful")
    print("Remaining balance:", balance)
except ATMError as e:
    print("ATM Error:", e)
```

project 1: student record management

```python
FILENAME = "student.txt"

def add_student():
    with open(FILENAME, "a") as file:
        sid = input("Enter Student ID: ")
        name = input("Enter Student Name: ")
        age = input("Enter Age: ")
        course = input("Enter Course: ")

        file.write(f"{sid},{name},{age},{course}\n")
        print("Student record added successfully")

def view_students():
    try:
        with open(FILENAME, "r") as file:
            records = file.readlines()
            if not records:
                print("No student records found")
                return

            print("\nStudent Records:")
            for record in records:
                sid, name, age, course = record.strip().split(",")
                print(f"ID: {sid}, Name: {name}, Age: {age}, Course: {course}")
    except FileNotFoundError:
        print("File not found")

def search_student():
    search_id = input("Enter Student ID to search: ")
    found = False

    try:
        with open(FILENAME, "r") as file:
            for record in file:
                sid, name, age, course = record.strip().split(",")
                if sid == search_id:
                    print("\nStudent Found:")
                    print(f"ID: {sid}")
                    print(f"Name: {name}")
                    print(f"Age: {age}")
                    print(f"Course: {course}")
                    found = True
                    break

        if not found:
            print("Student not found")
    except FileNotFoundError:
        print("File not found")

def main():
    while True:
        print("\n--- Student Record Management ---")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            print("Exiting program")
            break
        else:
            print("Invalid choice")

main()
```

project 2: word counter tool

```python
with open("summary.txt", "w") as file:
    file.write("Python is easy to learn\n")
    file.write("File handling is important\n")
    file.write("Exception handling makes code safe\n")

print("summary.txt file created")
```

```python
FILENAME = "summary.txt"

def word_counter():
    try:
        with open(FILENAME, "r") as file:
            lines = file.readlines()

            line_count = len(lines)
            word_count = 0
            char_count = 0

            for line in lines:
                word_count += len(line.split())
                char_count += len(line)

            print("Lines:", line_count)
            print("Words:", word_count)
            print("Characters:", char_count)

    except FileNotFoundError:
        print("Error: File not found")
    except PermissionError:
        print("Error: Permission denied while accessing file")
    except Exception as e:
        print("Unexpected error:", e)

word_counter()
```

```python
import csv

FILENAME = "Employee_Salary_Dataset.csv"

def analyze_salary():
    try:
        salaries = []

        with open(FILENAME, "r") as file:
            reader = csv.DictReader(file)

            for row in reader:
                salaries.append(float(row["Salary"]))

        if not salaries:
            print("No salary data found")
            return

        highest = max(salaries)
        lowest = min(salaries)
        average = sum(salaries) / len(salaries)

        print("Highest Salary:", highest)
        print("Lowest Salary:", lowest)
        print("Average Salary:", round(average, 2))

    except FileNotFoundError:
        print("Error: employee.csv file not found")
    except ValueError:
        print("Error: Salary column contains invalid data")
    except Exception as e:
        print("Unexpected error:", e)

analyze_salary()
```

project 4: Atm simulation

```python
class ATMError(Exception):
    pass


class ATM:
    def __init__(self, name, balance=5000):
        self.name = name
        self.balance = balance

    def check_balance(self):
        print("Current Balance:", self.balance)

    def deposit(self, amount):
        if amount <= 0:
            raise ATMError("Deposit amount must be positive")
        self.balance += amount
        print("Amount deposited successfully")

    def withdraw(self, amount):
        if amount <= 0:
            raise ATMError("Withdrawal amount must be positive")

        if amount % 100 != 0:
            raise ATMError("Amount must be in multiples of 100")

        if amount > self.balance:
            raise ATMError("Insufficient balance")

        if self.balance - amount < 1000:
            raise ATMError("Minimum balance of 1000 must be maintained")

        self.balance -= amount
        print("Please collect your cash")

def main():
    atm = ATM("User")

    while True:
        print("\n--- ATM MENU ---")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                atm.check_balance()

            elif choice == "2":
                amount = float(input("Enter deposit amount: "))
                atm.deposit(amount)

            elif choice == "3":
                amount = float(input("Enter withdrawal amount: "))
                atm.withdraw(amount)

            elif choice == "4":
                print("Thank you for using ATM")
                break

            else:
                print("Invalid choice")

        except ATMError as e:
            print("ATM Error:", e)

        except ValueError:
            print("Invalid input. Enter numeric value only")

main()
```

project 5: user login systems

```python
FILENAME = "user.txt"

def register_user():
    try:
        username = input("Enter new username: ")
        password = input("Enter new password: ")

        with open(FILENAME, "a") as file:
            file.write(f"{username},{password}\n")

        print("User registered successfully")

    except Exception as e:
        print("Error during registration:", e)


def login_user():
    try:
        username = input("Enter username: ")
        password = input("Enter password: ")

        found = False

        with open(FILENAME, "r") as file:
            for line in file:
                saved_user, saved_pass = line.strip().split(",")

                if username == saved_user and password == saved_pass:
                    found = True
                    break

        if found:
            print("Login successful")
        else:
            print("Invalid username or password")

    except FileNotFoundError:
        print("Error: user.txt file not found")
    except ValueError:
        print("Error: Invalid file format")
    except Exception as e:
        print("Unexpected error:", e)


def main():
    while True:
        print("\n--- USER LOGIN SYSTEM ---")
        print("1. Register")
        print("2. Login")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            register_user()
        elif choice == "2":
            login_user()
        elif choice == "3":
            print("Exiting program")
            break
        else:
            print("Invalid choice")

main()
```

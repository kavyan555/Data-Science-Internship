```python
# Task 1: Student Grade Calculator

def calculate_grade(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 75:
        return "A"
    elif average >= 60:
        return "B"
    elif average >= 40:
        return "C"
    else:
        return "Fail"


# Input: list of marks
marks_list = [78, 85, 69, 90, 88]

# Function call
grade = calculate_grade(marks_list)

print("Marks:", marks_list)
print("Grade:", grade)
```

```python
# Task 2: ATM Withdrawal Function

def withdraw(balance, amount):
    if amount > balance:
        return "Insufficient balance"
    elif amount % 100 != 0:
        return "Amount must be a multiple of 100"
    else:
        balance -= amount
        return balance


# Example usage
current_balance = 10000
withdraw_amount = 2500

result = withdraw(current_balance, withdraw_amount)

print("Result:", result)
```

```python
# Task 3: Password Validator Function

def validate_password(password):
    if len(password) < 8:
        return "Password must be at least 8 characters long"

    has_digit = False
    has_upper = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char.isupper():
            has_upper = True

    if not has_digit:
        return "Password must contain at least one digit"
    if not has_upper:
        return "Password must contain at least one uppercase letter"

    return "Password is valid"


# Example usage
password = "Hello123"
result = validate_password(password)
print(result)
```

```python
# Task 4: Employee Salary Calculator Function

def calculate_net_salary(basic_salary):
    hra = basic_salary * 0.20
    pf = basic_salary * 0.12
    net_salary = basic_salary + hra - pf
    return net_salary


# Example usage
basic_salary = 30000
net = calculate_net_salary(basic_salary)

print("Basic Salary:", basic_salary)
print("Net Salary:", net)
```

```python
# Task 5: File Data Counter Function

def count_file_data(filename):
    with open(filename, "r") as file:
        lines = file.readlines()

    line_count = len(lines)
    word_count = 0
    char_count = 0

    for line in lines:
        word_count += len(line.split())
        char_count += len(line)

    return line_count, word_count, char_count


# Example usage
file_name = "sample.txt"
lines, words, characters = count_file_data(file_name)

print("Number of Lines:", lines)
print("Number of Words:", words)
print("Number of Characters:", characters)
```

```python
#task 02:sort employees by salary(lambda)

employees = [
    {"name": "Shreya", "salary": 45000},
    {"name": "Suhas", "salary": 30000},
    {"name": "Swetha", "salary": 55000},
    {"name": "Arjun", "salary": 40000}
]

sorted_employees = sorted(employees, key=lambda x: x["salary"])

print("Employees Sorted by Salary:")
for emp in sorted_employees:
    print(emp["name"], "-", emp["salary"])
```

```python
#task 03: square & cube calculator(lambda+map)

numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x ** 2, numbers))
cubes = list(map(lambda x: x ** 3, numbers))

print("Squares:", squares)
print("Cubes:", cubes)
```

```python
#task 04: discount calculator

# Discount: 10% if amount >= 5000 else no discount

discount_calculator = lambda amount: amount - (amount * 0.10) if amount >= 5000 else amount

bill_amount = 6500
final_amount = discount_calculator(bill_amount)

print("Bill Amount:", bill_amount)
print("Final Amount after Discount:", final_amount)
```

```python
def even_or_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(7))
```

```python
def sum_to_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(sum_to_n(5))
```

```python
def find_max(numbers):
    maximum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
    return maximum

print(find_max([10, 45, 23, 89, 12]))
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print(fact)

factorial(5)
```

```python
def reverse_words(sentence):
    words = sentence.split()
    reversed_words = []

    for word in words:
        reversed_words.append(word[::-1])

    return " ".join(reversed_words)

print(reverse_words("Python is very easy"))
```

```python
def count_frequency(elements):
    freq = {}
    for item in elements:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    return freq

print(count_frequency([1, 2, 2, 3, 1, 4, 2]))
```

```python
def is_prime(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"
    return "Prime"

print(is_prime(7))
```

```python
import pandas as pd
import numpy as np

# STEP 1: Create Dataset
data = {
    "Customer_ID": [1,2,3,4,5,6,2,3],
    "Order_ID": [101,102,103,104,105,106,102,103],
    "Gender": ["M","F","m","f","Male","Female","M","F"],
    "Age": [25, -5, 200, 35, np.nan, 28, 30, 22],
    "City": ["Bangalore","Chennai",None,"Delhi","Delhi",None,"Chennai","Mumbai"],
    "Product_Name": ["Laptop","Mouse","Keyboard","Laptop","Tablet","Mouse","Keyboard","Laptop"],
    "Product_Category": ["Electronics","electronics","Electronics","ELEC","electronics","Electronics","ELEC","electronics"],
    "Quantity": [1,2,1,3,2,5,1,2],
    "Price": [50000,500,1500,50000,20000,500,1500,50000],
    "Order_Date": ["2024-01-05","2024-02-10","2024-03-15","2024-01-20","2024-04-01","2024-02-25","2024-03-15","2024-01-05"]
}

df = pd.DataFrame(data)

print("DATA BEFORE CLEANING:\n")
print(df)
print("\n")

# STEP 2: Remove duplicates (IMPORTANT .copy() to avoid warnings)
df = df.drop_duplicates(subset="Order_ID").copy()

# STEP 3: Fix Gender
df["Gender"] = df["Gender"].str.lower()
df["Gender"] = df["Gender"].replace({"m":"male","f":"female"})

# STEP 4: Handle Age Issues
df.loc[(df["Age"] < 0) | (df["Age"] > 100), "Age"] = np.nan
df["Age"].fillna(df["Age"].median(), inplace=True)

# STEP 5: Fill Missing City using Mode
df["City"].fillna(df["City"].mode()[0], inplace=True)

# STEP 6: Standardize Category
df["Product_Category"] = df["Product_Category"].str.lower()
df["Product_Category"] = df["Product_Category"].replace({"elec":"electronics"})

# STEP 7: Convert to Datetime
df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors="coerce")

# STEP 8: Extract Month & Day
df["Month"] = df["Order_Date"].dt.month_name()
df["Day"] = df["Order_Date"].dt.day_name()

# STEP 9: Rename columns to lowercase
df.columns = df.columns.str.lower()

# STEP 10: Total Revenue column
df["total_price"] = df["quantity"] * df["price"]

# STEP 11: Business Insights
total_revenue = df["total_price"].sum()
top_cities = df.groupby("city")["total_price"].sum().sort_values(ascending=False)
category_sales = df.groupby("product_category")["total_price"].sum()

# STEP 12: Display Results
print("Cleaned Data:\n")
print(df)

print("\nTotal Revenue:", total_revenue)

print("\nTop Cities by Sales:\n")
print(top_cities)

print("\nCategory Wise Sales:\n")
print(category_sales)
```

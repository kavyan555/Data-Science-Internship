1. Stock Price Movement

Best Chart: Line Plot

Because stock prices change over time â†’ shows trend.

```python
import matplotlib.pyplot as plt
days = ["Mon","Tue","Wed","Thu","Fri"]
price = [100,105,102,110,115]

plt.plot(days, price, marker='o')
plt.title("Stock Price Movement")
plt.show()
```

2. Website Traffic Trend

Best Chart: Line Plot

 Used to show daily/monthly visitors trend.

```python
days = ["Mon","Tue","Wed","Thu","Fri"]
visitors = [200,350,300,450,500]

plt.plot(days, visitors)
plt.title("Website Traffic Trend")
plt.show()
```

3. Employee Salary Distribution

Best Chart: Histogram

Shows how salaries are distributed.

```python
import numpy as np
salary = np.random.randint(20000,100000,100)

plt.hist(salary, bins=20)
plt.title("Salary Distribution")
plt.show()
```

4. Loan Default Rate

Best Chart: Pie Chart

Shows percentage of defaulters vs non-defaulters.

```python
labels = ["Default","No Default"]
values = [30,70]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.show()
```

5. Exam Marks Analysis

10. Sales Forecast vs Actual

Best Chart: Multiple Line Plot

Compare predicted vs real sales.

```python
months = ["Jan","Feb","Mar"]
actual = [1000,1200,1500]
forecast = [900,1300,1400]

plt.plot(months, actual, label="Actual")
plt.plot(months, forecast, label="Forecast")
plt.legend()
plt.show()
```
